# üîπ Complex Data Type in Python

### 1. **Definition**

* A complex number is written as:
  [
  z = a + bj
  ]
  where:

  * `a` ‚Üí real part
  * `b` ‚Üí imaginary part
  * `j` ‚Üí imaginary unit (`j^2 = -1`)

‚ö†Ô∏è Note: Python uses **`j`**, not `i`.

---

### 2. **Creation of Complex Numbers**

```python
z1 = 3 + 4j        # directly using literal
z2 = complex(3, 4) # using constructor
print(z1)          # (3+4j)
print(z2)          # (3+4j)
```

---

### 3. **Attributes**

```python
z = 5 + 6j
print(z.real)   # 5.0
print(z.imag)   # 6.0
print(z.conjugate())  # (5-6j)
```

---

### 4. **Basic Operations**

```python
z1 = 2 + 3j
z2 = 4 + 5j

print(z1 + z2)  # (6+8j)
print(z1 - z2)  # (-2-2j)
print(z1 * z2)  # (-7+22j)
print(z1 / z2)  # (0.5609756097560976+0.0487804878048781j)
```

---

### 5. **Magnitude (Modulus)**

```python
z = 3 + 4j
print(abs(z))  # 5.0  (since sqrt(3^2 + 4^2))
```

---

### 6. **cmath Module (Advanced Complex Math)**

Python‚Äôs `cmath` library is made for complex numbers.

```python
import cmath

z = 3 + 4j

# Polar form
print(cmath.polar(z))  # (5.0, 0.927...) -> (r, Œ∏)

# Rectangular form from polar
print(cmath.rect(5, cmath.pi/4))  # (3.5355339059327378+3.5355339059327378j)

# Phase angle
print(cmath.phase(z))  # 0.927... (radians)
```

---

### 7. **Conversion Rules**

Python can cast to/from complex:

```python
print(complex(5))        # (5+0j)
print(complex(2, -3))    # (2-3j)
print(complex("7+2j"))   # (7+2j)
```

‚ö†Ô∏è Strings must be in the correct format, otherwise error.

---

### 8. **Important Points**

* `complex` is a built-in data type.
* `isinstance(z, complex)` ‚Üí check if a variable is complex.
* Real numbers are treated as `complex` with imaginary part = `0`.
* No automatic implicit conversion from `complex` ‚Üí `float` or `int` (must convert manually if imaginary part is zero).

Example:

```python
z = 5 + 0j
print(float(z.real))  # 5.0
```

## üîπ 1. **f-strings (Python 3.6+)**

* Embed variables or expressions inside `{}`.
* Syntax: `f"text {expression}"`

### Examples:

```python
name = "Alice"
age = 25
pi = 3.14159265

# Simple variable
print(f"My name is {name} and I am {age} years old.")
# Output: My name is Alice and I am 25 years old.

# Expression inside {}
print(f"Next year I will be {age + 1}")
# Output: Next year I will be 26

# Number formatting
print(f"Pi rounded to 2 decimals: {pi:.2f}")
# Output: Pi rounded to 2 decimals: 3.14

# Alignment and padding
print(f"|{name:<10}|{age:^5}|")  # left-align name, center-align age
# Output: |Alice     | 25  |
```

---

## üîπ 2. **`format()` Method**

* Works in all Python 3 versions.
* Can use **positional** or **keyword arguments**.

### Examples:

```python
name = "Bob"
age = 30
pi = 3.14159265

# Positional
print("Name: {}, Age: {}".format(name, age))
# Output: Name: Bob, Age: 30

# Keyword
print("Name: {n}, Age: {a}".format(n=name, a=age))
# Output: Name: Bob, Age: 30

# Number formatting
print("Pi: {:.3f}".format(pi))
# Output: Pi: 3.142

# Alignment and padding
print("|{:<10}|{:^5}|".format(name, age))  # left-align, center-align
# Output: |Bob       | 30  |
```

---

### üîπ Key Points

1. **f-strings** are preferred: cleaner, faster, and allow expressions.
2. **`format()`** works in all Python 3 versions; useful for older code.
3. **Number formatting:** `{variable:.2f}` ‚Üí rounds float to 2 decimals.
4. **Alignment & padding:** `<` left, `>` right, `^` center.

## üîπ Docstring in Python (Simple Notes)

* **Definition:** A **docstring** is a string literal used to **document a function, class, or module**.
* **Syntax:** Use **triple quotes `""" ... """`** right after the definition.

### üîπ Examples:

**Function:**

```python
def add(a, b):
    """Return sum of a and b."""
    return a + b
```

**Class:**

```python
class Car:
    """Represents a car."""
    def __init__(self, brand, model):
        """Initialize brand and model."""
        self.brand = brand
        self.model = model
```

**Module:**

```python
"""This module contains math functions."""
```

### üîπ Accessing Docstrings:

```python
print(add.__doc__)       # "Return sum of a and b."
help(Car)                # Shows class docstring
```

### üîπ Key Points:

* Stored in **`.__doc__`** attribute.
* Helps **readability** and **automatic documentation**.
* Preferred over comments for explaining **what a function/class/module does**.
